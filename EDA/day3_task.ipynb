Matplotlib

```python
#Line plot

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2.5, 5, 7.5, 10.5, 12.5]
plt.plot(x, y, color='green', marker='o')
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#Scatter plot

import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

df = pd.DataFrame({
"Height": [150, 160, 170, 180, 190],
"Weight": [50, 60, 70, 80, 90],
"Age": [20, 25, 30, 35, 40]
})
scatter_matrix(df, alpha=0.8, figsize=(6, 4), diagonal="hist")
plt.suptitle("Scatter Matrix Example", size=12)
plt.show()

```

```python
#Lag plot

import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import lag_plot

data = pd.Series([10, 12, 16, 21, 22, 25, 30, 34, 40, 42, 48])
plt.figure(figsize=(6, 5))
lag_plot(data)
plt.title("Lag Plot Example")
plt.show()
```

```python
#Autocorrelation plot

import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import autocorrelation_plot

ts = pd.Series([10, 12, 15, 20, 22, 25, 30, 35, 40, 42, 45])
autocorrelation_plot(ts)
plt.title("Autocorrelation Plot")
plt.show()
```

```python
#Bootstrap plot

import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import bootstrap_plot

ts = pd.Series([10, 12, 15, 20, 22, 25, 30, 35, 40, 42, 45])
bootstrap_plot(ts, size=7, samples=500, color='pink')
plt.show()
```

```python
#Bar chart

import matplotlib.pyplot as plt

labels = ['Rent', 'Food', 'Savings', 'Entertainment']
sizes = [40, 30, 20, 10]
colors = ['blue','purple','pink','green']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Monthly Expense Breakdown')
plt.axis('equal')
plt.show()
```

```python
#Bar chart

import matplotlib.pyplot as plt

categories = ['January', 'February', 'March', 'April']
values = [10, 24, 36, 40]
plt.bar(categories, values, color='skyblue', edgecolor='navy')
plt.title('Monthly Sales Growth')
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.show()
```
