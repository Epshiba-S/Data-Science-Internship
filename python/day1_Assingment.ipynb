Hangman Game

```python

import random

words = ["python", "apple", "computer", "science"]
word = random.choice(words)

guessed = ["_"] * len(word)

stages = [
    """
       |
       |
       |
    =====
    """,
    """
       |
       O
       |
    =====
    """,
    """
       |
       O
      /|
    =====
    """,
    """
       |
       O
      /|\\
    =====
    """,
    """
       |
       O
      /|\\
      /
    =====
    """,
    """
       |
       O
      /|\\
      / \\
    =====
    """
]

wrong_attempts = 0
max_attempts = 6
guessed_letters = []

while wrong_attempts < max_attempts and "_" in guessed:
    print(stages[wrong_attempts])
    print("Word:", " ".join(guessed))

    guess = input("Enter a letter: ").lower()

    if len(guess) != 1 or not guess.isalpha():
        print("Invalid input\n")
        continue

    if guess in guessed_letters:
        print("Already guessed\n")
        continue

    guessed_letters.append(guess)

    if guess in word:
        for i in range(len(word)):
            if word[i] == guess:
                guessed[i] = guess
    else:
        wrong_attempts += 1
        print("Wrong guess!\n")

# Result
if "_" not in guessed:
    print("You won! Word:", word)
else:
    print(stages[wrong_attempts - 1])
    print("Game Over! Word:", word)
```
