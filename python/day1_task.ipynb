```python
# 1. number of character
s="hello"
print(len(s))
```

```python
# 2. Reverse string
s="hello"
print(s[::-1])
```

```python
# 3. palindrome check
s="madam"
if (s==s[::-1]):
  print("s is palindrome")
else:
  print("s is not palindrome")
```

```python
# 4. vowels&consonants
s="hello"
vowels="aeiou"
v=c=0
for ch in s:
  if ch in vowels:
    v+=1
  else:
    c+=1
print("Vowels:",v,"Consonants:",c)
```

```python
# 5. uppercase&lowercase
s="Hello"
print(s.upper())
print(s.lower())
```

```python
# 6. remove space
s="hello world"
print(s.replace(" ",""))
```

```python
# 7. replace word
s=" I like apple"
print(s.replace("apple","mango"))
```

```python
# 8. char frequency
s="hello"
freq={}
for ch in s:
  freq[ch]=freq.get(ch,0)+1
print(freq)
```

```python
# 9. extract substring
s="hello world"
print(s[0:5])
```

```python
# 10. anagram check
s1="listen"
s2="silent"
if (sorted(s1)==sorted(s2)):
  print("s1 and s2 are anagram")
else:
  print("s1 and s2 are not anagram")
```

```python
# 2. add&remove
l=[1,2,3]
l.append(4)
l.remove(3)
print(l)
```

```python
# 3. max&min
l=[6,8,5]
print("Max:",max(l))
print("Min:",min(l))
```

```python
# 4. reverse list
l=[1,3,2]
print("Reverse list:",l[::-1])
```

```python
# 5. sort
l=[1,2,3]
print(sorted(l))
print(sorted(l,reverse=True))
```

```python
# 6. count occurrences
l=['apple','mango','grapes','mango','orange','pineapple','mango']
print("Total mango:",l.count('mango'))
print("Total apple:",l.count('apple'))
```

```python
# 7. merge lists
list1=[18,20]
list2=[19,21]
print(list1+list2)
```

```python
# 8. remove duplicates
l=[1,2,3,2,4]
print(list(set(l)))
```

```python
# 9. second largest
l=[10,20,30,70]
l.sort()
print("Second largest:",l[-2])
```

```python
# 10. even&odd split
l=[1,2,3,4]
even=[x for x in l if x%2==0]
print(even)
odd=[x for x in l if x%2!=0]
print(odd)
```
