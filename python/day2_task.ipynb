```python
#1.
try:
  user1=int(input("Enter a number1:"))
  user2=int(input("Enter a number2:"))
  print("User value:",user1+user2)
except ValueError:
  print("Invalid input!")
```

```python
#2.
try:
  a=int(input("Enter a:"))
  b=int(input("Enter b:"))
  print("The Result:",a/b)
except ZeroDivisionError:
  print("Cannot divide by zero")
except ValueError:
  print("Invalid input!")
```

```python
#3.
l=[21,11,51,41,71]
try:
  index=int(input("Enter index:"))
  print(l[index])
except IndexError:
  print("Index out ofrange")
```

```python
#4.
try:
  f=open("file.txt","r")
  content=f.read()
  print(content)
  f.close()
except FileNotFoundError:
  print("File not found")
```

```python
#5.
class LoginError(Exception):
  pass
correct="admin123"
attempts=3
while attempts>0:
  pwd=input("Enter password:")
  if pwd==correct:
    print("Login sucessful")
    break
  attempts-=1
if attempts==0:
  raise LoginError("Too many attempts!")
```

```python
#6.
try:
  num=float(input("Enter a number:"))
  print(num)
except ValueError:
  print("Invalid float")
```

```python
#7.
balance=2500
try:
  amount=int(input("Enter the amount:"))
  if amount>balance:
    raise ValueError("Insufficient balance")
  else:
    print("Transaction sucessfully!")
    balance-=amount
    print("Remaining balance:",balance)
except  Exception as e:
  print(e)
```

```python
#8.
try:
  age=int(input("Enter age:"))
  if age<0:
    raise ValueError("Invalid age")
  else:
    print("Valid age")
except ValueError as e:
  print(e)
```

```python
#2.
file = open("users.txt", "r")

word = input("Enter word: ")

for line in file:
    if word in line:
        print(line)

file.close()
```

```python
#3.
file = open("users.txt", "r")

content = file.read()

words = len(content.split())
characters = len(content)

file.seek(0)
lines = len(file.readlines())

print("Words:", words)
print("Lines:", lines)
print("Characters:", characters)

file.close()
```

```python
#7.
file = open("users.txt", "r")

lines = file.readlines()

unique = []

for line in lines:
    if line not in unique:
        unique.append(line)

file.close()

file = open("data.txt", "w")

for line in unique:
    file.write(line)

file.close()

print("Duplicate lines removed")
```

```python
#8.
file = open("marks.txt", "w")

total = 0

for i in range(3):
    mark = int(input("Enter mark: "))

    total = total + mark

    file.write(str(mark) + "\n")

average = total / 3

print("Total Marks:", total)
print("Average Marks:", average)

file.close()
```

```python
#10.
file = open("users.txt", "r")

data = file.read()

file.close()

old_name = input("Enter old name: ")
new_name = input("Enter new name: ")

data = data.replace(old_name, new_name)

file = open("users.txt", "w")

file.write(data)

file.close()

print("Record updated")
```

```python
#6.
def clear_whitespace(text):
    return text.strip()

def to_uppercase(text):
    return text.upper()

raw_string = "   hello world   "


clean_text = clear_whitespace(raw_string)
final_text = to_uppercase(clean_text)


print(f"Original: '{raw_string}'")
print(f"Cleaned:  '{clean_text}'")
print(f"Final:    '{final_text}'")
```
